```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
df = pd.read_csv("data/stats03.csv")
z = df.to_numpy()
```

```python
plt.hist(z, bins=12)
plt.xlabel("")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(z, bins=12, density=True)
plt.xlabel("")
plt.ylabel("Probability")
plt.show()
```

#### 期待値、分散、歪度、尖度、平均中心の５次モーメント

```python
print("期待値：{:.3f}".format(np.mean(z)))
print("分散：{:.3f}".format(np.var(z)))
print("歪度: {:.3f}".format(ss.skew(z)[0]))
print("尖度: {:.3f}".format(ss.kurtosis(z)[0]))
print("平均中心の5次モーメント: {:.3f}".format(ss.moment(z, moment=5)[0]))
```
