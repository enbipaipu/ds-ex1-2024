```python
import pandas as pd
```

```python
df = pd.read_csv('data/score.csv')
display(df)
```

```python
df_class1 = df[df["class"] == 1]
display(df_class1.head())
```

```python
ser = pd.Series([170,3,80,75],index=df.columns)

df = pd.concat([df, ser.to_frame().T],ignore_index=True)
display(df)
```

```python
df["ave"] = (df["score1"] + df["score2"]) / 2
display(df.head())
```

```python
df_sort = df.sort_values(by="ave", ascending=False)
display(df_sort.head())
```

```python
df_agg = df.groupby("class").mean(numeric_only=True)
df_agg2 = df_agg.drop(columns="id")
display(df_agg2)
```
