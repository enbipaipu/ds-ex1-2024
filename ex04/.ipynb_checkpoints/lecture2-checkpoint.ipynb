```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = np.random.binomial(n=100, p=0.8, size=1000)
plt.hist(data1)
plt.show()
```

```python
data2 = np.random.poisson(lam=10, size=1000)
plt.hist(data2)
plt.show()
```

```python
data3 = np.random.randn(1000)
plt.hist(data3)
plt.show()
```

```python
data4= np.random.normal(loc=-2,scale=1.4,size=1000)
plt.hist(data4)
plt.show()
```

```python
data5 = np.random.exponential(scale=1/1.3, size=1000)
plt.hist(data5)
plt.show()
```

```python
data6 = np.random.rand(1000)*5
plt.hist(data6)
plt.show()
```
