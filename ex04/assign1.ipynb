### 3,numpyのランダム生成でヒストグラムを書く

```python
date= np.random.poisson(lam=5.0, size=2000)
plt.hist(date, bins=14)
plt.xlabel("x")
plt.ylabel("Frequency")
plt.show()

x = [x for x in range(0, 15)]
print(x)
date2 = [ss.poisson.pmf(k=t, mu=5) for t in x]
plt.plot(x, date2)
plt.xlabel("x")
plt.ylabel("PMF")
plt.show()
```
