```python
import pandas as pd
import matplotlib.pyplot as plt
```

## データの読み込み

```python
df_D = pd.read_csv('SSDSE/D/SSDSE-D-2023.csv', skiprows=1, encoding="shift-jis")
df_D1 = df_D[df_D["男女の別"]=="0_総数"]
df_D1 = df_D1[df_D1["都道府県"]!="全国"]
display(df_D1.head())
print(df_D1.shape)
```

```python
df_D2=df_D1["推定人口（10歳以上）"]
print(df_D2)
```

### 標準化を行う関数

```python
def Zscore(lis):
    Expect = lis.mean()
    Std = lis.std()
    z = [(x-Expect)/Std for x in lis]
    return z
```

```python
plt.bar(Zscore(df_D2), df_D2, width=0.34,ec='black' )
plt.xlabel("Rondom variable")
plt.ylabel("Probability")
plt.show()
```
