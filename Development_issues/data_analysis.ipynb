```python
import pandas as pd
```

## データの読み込み

```python
df_D = pd.read_csv('SSDSE/D/SSDSE-D-2023.csv', skiprows=1, encoding="shift-jis")
display(df_D.head())
```

```python
df_D2 = df_D[df_D["都道府県"] != "全国"]
df_D2 = df_D2.groupby("都道府県").mean(numeric_only=True)
display(df_D2.sort_values(by="学習・自己啓発・訓練の総数",ascending=False))
print(df_D2.shape)
```
